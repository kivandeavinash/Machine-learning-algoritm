Sales prediction model

```python
#importing labraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the data set
data = pd.read_csv("/content/sls.csv")
data.head()
```

```python
#intialiating the variables
x = data.drop(['Purchased'], axis= 1)
y = data['Purchased']. values.reshape(-1,1)
```

Splitting data into trainig, testing dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
#x_test
```

```python
#training our model using logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
#testing our model
```

```python
y_pred = classifier.predict(x_test)
```

Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

prediction for new values

```python
#taking the input from the user
print("Enter the amount you will invest on: ")
age = float(input("Age : "))
estimetedsalary = float(input("Estimetedsalary"))

#predicting the purchesed with respect to the inputs
output = classifier.predict([[age,estimetedsalary]])
print(" Estimetedsalary on age base"\
      .format(output[0] if output else "not predictable",age,estimetedsalary))
```
